```python
train = pd.read_csv("input/titanic/train.csv")
test = pd.read_csv("input/titanic/test.csv")
```

```python
train.describe()
test.describe()
```

```python
train.head()
test.head()
```

```python
train=train.drop(columns=["Cabin", "Ticket"])
test=test.drop(columns=["Cabin", "Ticket"])
```

```python
train = train.dropna(axis=0)
test = test.dropna(axis=0)
```

```python
train.head()
test.head()
```

```python
train.describe()
test.describe()
```

```python
Onehot = pd.get_dummies(train['Embarked'])

train = pd.concat([train, Onehot], axis=1)
```

```python
Onehot = pd.get_dummies(test['Embarked'])

test = pd.concat([test, Onehot], axis=1)
```

```python
train.head()
```

```python
train.shape
```

```python
Onehot = pd.get_dummies(train['Sex'])
train = pd.concat([train, Onehot], axis=1)
```

```python
Onehot = pd.get_dummies(test['Sex'])
test = pd.concat([test, Onehot], axis=1)
```

```python
train = train.drop(columns=["Embarked", "Sex", "Name"])
train.head()
```

```python
test = test.drop(columns=["Embarked", "Sex", "Name"])
test.head()
```

```python

y = train.Survived
```

```python
train.drop(['Survived'], axis=1, inplace=True)
```

```python
train.head()
```

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(train, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
from xgboost.sklearn import XGBClassifier  


my_model_1 = XGBClassifier(random_state=1, learning_rate=0.5, max_depth=3)

my_model_1.fit(X_train_full, y_train)
```

```python
from sklearn.metrics import accuracy_score
predictions_1 = my_model_1.predict(X_valid_full)
```

```python
predictions_1
```

```python
accuracy = accuracy_score(y_valid, predictions_1)
print(accuracy)
```

```python
test_predictions = my_model_1.predict(test)
```

```python
test_predictions
```

```python
data = np.array(test_predictions)
data_series = pd.Series(data)
test = pd.concat([test, data_series], axis=1)
```
